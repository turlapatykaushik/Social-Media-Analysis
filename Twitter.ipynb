# WORLD GAMES

```python
import glob
import pandas as pd

df = pd.concat(map(pd.read_csv, glob.glob('D:\AISTS\Course Related\Research Paper\Federations Data\The World Games - IWGA\Twitter\*.csv')))
```

```python
df.corr()
```

# SORTING ON IMPRESSIONS

```python
df = df.sort_values(by='impressions')
print(df.shape)
df[400:]
```

# SORTING ON ENGAGEMENTS

```python
df = df.sort_values(by='engagements')
print(df.shape)
df[400:]
```

# SORTING ON HASHTAG CLICKS

```python
df = df[['Tweet permalink','Tweet text','time','hashtag clicks']].sort_values(by='hashtag clicks')
print(df.shape)
df[400:]
```
